CODE FOR PREDICTING TIME FOR BREAK

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv(r'C:\Users\LENOVO\Downloads\Book1.csv')
data
```

```python
x=data.drop("BREAKS IN MINS /HOUR OF STUDY",axis=1)
x
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder() 
x['Period']=label_encoder.fit_transform(x['Period'])
x['Period'].unique()
x
```

```python
y=data['BREAKS IN MINS /HOUR OF STUDY']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
RDF=RandomForestClassifier()
RDF.fit(x,y)
```

```python
RDF.score(x,y)
```

```python
RDF.predict([[1,5,20]])
```

```python
print(RDF.predict([[1,5,20]]))
```
